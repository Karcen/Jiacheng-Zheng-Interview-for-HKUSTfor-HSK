```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Garamond'
plt.rcParams['font.style'] = 'normal'


# CSV
df = pd.read_csv(r'C:\Users\Jiacheng Zheng\Desktop\ICM\ICM2024\Q1\Q1.csv')

# 
months = df['Month']
erie = df['Erie']
michigan = df['Michigan']
ontario = df['Ontario']
clair = df['Clair']
superior = df['Superior']

bar_width = 0.15  # 
index = np.arange(len(months))  # 

# 
colors = ['#1f78b4', '#33a02c', '#e31a1c', '#ff7f00', '#6a3d9a']
print(colors)


# 
plt.figure(figsize=(12, 6))
handles = []  # 
for i, (lake, month, color) in enumerate(zip([erie, michigan, ontario, clair, superior], months, colors)):
    bars = plt.bar(index + i * bar_width, lake, width=bar_width, label=f'{lake.name} - {month}', color=color)
    handles.append(bars[0])  # barhandles
    
    # # 
    # for bar in bars:
    #     plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
    #              f'{bar.get_height():.2f}', ha='center', va='bottom')

# 
plt.xlabel('Month',fontsize=20)
plt.ylabel('Water Level(m)',fontsize=20)

# 
plt.xticks(index + 2 * bar_width, months,fontsize=20)
plt.yticks(fontsize=15)
# 
legend_labels = [f'{lake.name} ' for lake, month in zip([erie, michigan, ontario, clair, superior], months)]
plt.legend(handles, legend_labels, bbox_to_anchor=(1.05, 1), loc='upper left',fontsize=20)
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\ICM\ICM2024\Q1\Q1.svg', format='svg', bbox_inches='tight')
# 
plt.show()
```
