```python
import matplotlib.pyplot as plt
import random

# 10[20, 100]
radii = [random.randint(20, 100) for _ in range(10)]

# 
colors = ['red', 'green', 'blue', 'purple']

# 
fig, ax = plt.subplots()
ax.set_aspect('equal', 'box')  # 

# 
for radius in radii:
    color = random.choice(colors)  # 
    circle = plt.Circle((0, 0), radius, color=color, fill=False, linewidth=2)
    ax.add_patch(circle)

# 
ax.set_xlim(-110, 110)
ax.set_ylim(-110, 110)

# 
ax.axis('off')

# 
plt.show()
```
