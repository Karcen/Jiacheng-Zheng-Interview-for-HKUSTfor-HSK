```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

# sin(x)
x = np.linspace(0, 4 * np.pi, 1000)
y = np.sin(x)

# 
fig = plt.figure(figsize=(10, 6))
gs = gridspec.GridSpec(2, 1, height_ratios=[2, 1])

# sin(x)
ax1 = plt.subplot(gs[0])
ax1.plot(x, y, label='sin(x)', color='blue')
ax1.set_title('sin(x) Function')

# 
start_idx = 200
end_idx = 400

# 
ax2 = plt.subplot(gs[1])
ax2.plot(x, y, label='sin(x)', color='blue')
ax2.set_title('Zoomed-in Area')

# 
ax2.set_xlim(x[start_idx], x[end_idx])
ax2.set_ylim(ax1.get_ylim())

# 
ax1.add_patch(plt.Rectangle((x[start_idx], ax1.get_ylim()[0]), x[end_idx] - x[start_idx], ax1.get_ylim()[1] - ax1.get_ylim()[0],
                            linewidth=1, edgecolor='r', facecolor='none'))

# 
plt.tight_layout()

# 
plt.show()
```
