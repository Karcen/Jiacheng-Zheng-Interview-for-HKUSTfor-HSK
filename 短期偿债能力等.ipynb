```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Garamond'
# Given data
years = ['2023', '2024', '2025', '2026']
current_ratio = [1.81, 2.43, 2.62, 2.77]
net_profit_margin = [1.53, 1.65, 1.77, 1.83]

# Create a line chart
plt.figure(figsize=(10, 5))
plt.plot(years, current_ratio, marker='o', label='Current Ratio')
plt.plot(years, net_profit_margin, marker='o', label='Net Profit Margin')
plt.title('Financial Ratios Over Years')
plt.xlabel('Year')
plt.ylabel('Ratio Values')
plt.legend()

# Add data labels to the points
for i, ratio in enumerate(current_ratio):
    plt.annotate(f'{ratio:.2f}', (years[i], ratio), textcoords="offset points", xytext=(0, 5), ha='center', color='blue')

for i, ratio in enumerate(net_profit_margin):
    plt.annotate(f'{ratio:.2f}', (years[i], ratio), textcoords="offset points", xytext=(0, 5), ha='center', color='orange')

# Show or save the chart
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\financial_ratios_chart.png', dpi=1200)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Garamond'

# 
years = [2023, 2024, 2025]

# 
profit_margin = [43.68, 46.79, 56.54]

# 
roe = [10.05, 11.29, 17.44]

# 
cost_ratio = [77.57, 87.93, 129.62]

# 
plt.figure(figsize=(10, 6))

# 
plt.plot(years, profit_margin, label='Profit Margin', marker='o')
plt.plot(years, roe, label='ROE', marker='o')
plt.plot(years, cost_ratio, label='Cost Ratio', marker='o')

# 
for i in range(len(years)):
    plt.text(years[i], profit_margin[i], f'{profit_margin[i]:.2f}%', ha='right')
    plt.text(years[i], roe[i], f'{roe[i]:.2f}%', ha='right')
    plt.text(years[i], cost_ratio[i], f'{cost_ratio[i]:.2f}%', ha='right')

# 
plt.xticks(years)

# 
plt.title('Financial Ratios Over Years')
plt.xlabel('Year')
plt.ylabel('Percentage')

# 
plt.legend()
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Financial Ratios Over Years.png', dpi=1200)
# 
plt.show()
```
