```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Garamond'
# 
indicators = ['Emotional', 'Self', 'Interpersonal', 'Cognitive', 'Adaptation']

# 
company_scores = [445, 450, 440, 460, 475]
technical_scores = [450, 460, 455, 452, 480]
management_scores = [525, 545, 525, 495, 550]
civil_servant_scores = [460, 470, 460, 455, 490]
worker_scores = [455, 450, 390, 385, 475]
unemployed_scores = [360, 353, 351, 410, 380]

# 
marker_styles = ['o', '^', 's', 'd', 'v', 'h']

# 
plt.figure(figsize=(6, 3))  # 
plt.plot(indicators, company_scores, marker=marker_styles[0])
plt.plot(indicators, technical_scores, marker=marker_styles[1])
plt.plot(indicators, management_scores, marker=marker_styles[2])
plt.plot(indicators, civil_servant_scores, marker=marker_styles[3])
plt.plot(indicators, worker_scores, marker=marker_styles[4])
plt.plot(indicators, unemployed_scores, marker=marker_styles[5])

# 
plt.legend(['Company Employees', 'Technical Professionals', 'Managers', 'Civil Servants', 'Factory Workers', 'Unemployed'], loc='center left', bbox_to_anchor=(1, 0.5))

# 
plt.xlabel('Indicators')

# 
plt.ylabel('Scores')

# 
plt.title('China Mental Health Scale Dimension Scores of Different Occupations')

# 
plt.yticks(np.arange(350, 600, 50))
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Re-draw\China Mental Health Scale Dimension Scores of Different Occupations_en.png', dpi=1200, bbox_inches = 'tight')
# 
plt.show()
```
