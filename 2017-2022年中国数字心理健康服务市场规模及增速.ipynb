```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = 'Garamond'
# 
years = [2017, 2018, 2019, 2020, 2021, 2022]
market_size = [5, 7.5, 11, 15, 23.2, 33.8]
growth_rate = [0, 50, 46.67, 36.36, 54.67, 45.69]

# 
years_label = ['2017', '2018', '2019', '2020', '2021', '2022']
market_size_label = 'Market Size (billion yuan)'
growth_rate_label = 'Growth Rate (%)'

# 
ind = np.arange(len(years))

# y
fig, ax1 = plt.subplots(figsize=(10, 6))
ax2 = ax1.twinx()

# 
bar_width = 0.4
ax1.bar(ind, market_size, bar_width, color='#ECAD9E', label=market_size_label)

# 
ax1.legend(loc='upper left')
ax1.set_xlabel('Years')
ax1.set_ylabel(market_size_label)

# 
ax2.plot(ind, growth_rate, color='blue', marker='o', linestyle='-', linewidth=2, markersize=8, label=growth_rate_label)

# y
ax2.set_ylabel(growth_rate_label)

# 
plt.xticks(ind, years_label)

# 
ax2.legend(loc='upper right')

# 
plt.title('Market size and growth rate of digital mental health services in China, 2017-2022')

plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Re-draw\Market size and growth rate of digital mental health services in China, 2017-2022_en.png', dpi=1200, bbox_inches = 'tight')
# 
plt.tight_layout()
plt.show()
```
